```python
import pandas as pd
```

```python
path = "./SPOTS"

import os

path_list= []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".txt"):
            path_list.append(os.path.join(root, file))
            
path_list  
```

```python
col = ['Name', 'Missing', 'Centroid X µm', 'Centroid Y µm', 'Unique ID', 'TMA']
df_SPOTS = pd.DataFrame([],columns=col)
print(df_SPOTS.shape)
for file in path_list:
    df = pd.read_csv(file, encoding='iso-8859-1', delimiter='\t', dtype = {'Unique ID': str})
    df['TMA']= file
    df_SPOTS = df_SPOTS.append(df)
    
# start stop and step variables 
start, stop, step = -12, -4, 1
df_SPOTS['TMA'] = df_SPOTS['TMA'].str.slice(start,stop,step)
```

```python
df_SPOTS.to_csv('spot_list_20_02.csv',sep='\t', columns=['TMA','Name', 'Missing', 'Centroid X µm', 'Centroid Y µm', 'Unique ID' ])
```
