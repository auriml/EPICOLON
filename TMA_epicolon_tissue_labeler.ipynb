```python
def load_tiles(spot_list, level = 1, size = (75,75), overlap = 0.8, suffix = 'MMR'):
    lev = level
    try:
        p_ds = f'{suffix}_tiles_level_{lev}_size_{size[0]}_overl_{overlap}.csv'
        ds = pd.read_csv(p_ds)
        #ds = pd.concat([load_dataset(0)])
        ds['tile']=ds['tile'].apply(eval)
    except:
        print('failed')
        res = parallel(partial(process_spot, overlap = overlap, size = size, level = level, skip_scarce_tissue = False),spot_list)
        ds = pd.concat((*res,))
        ds.to_csv(p_ds,index=False)
    return ds
```
